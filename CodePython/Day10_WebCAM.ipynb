```python
import cv2

def show_webcam(mirror=False):
    cam = cv2.VideoCapture(0)
    while True:
        ret_val, img = cam.read()
        if mirror:            
            img = cv2.flip(img, 1)
            
        img = cv2.Canny(img,100,200)  
        print(img)
        cv2.imshow('my webcam', img[:,:,2])
        if cv2.waitKey(1) == 27: 
            break  # esc to quit
    cv2.destroyAllWindows()


def main():
    show_webcam(mirror=False)

if __name__ == '__main__':
    main()
```
