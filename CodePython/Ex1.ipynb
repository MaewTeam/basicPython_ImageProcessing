```python
img_name = "p1.png"
print(img_name)
```

```python
image_in = cv2.imread(img_name)
rgb = cv2.cvtColor(image_in, cv2.COLOR_BGR2RGB)
#print(image_in)
```

```python
b = rgb.copy()
```

```python
b[:, :, 0] = 0
b[:, :, 1] = 0
```

```python
cv2.imshow('RED image',b)
```

```python
plt.subplot(1, 2, 1), plt.imshow(rgb, 'gray')
plt.subplot(1, 2, 2), plt.imshow(b, 'gray')
plt.show()

cv2.waitKey(0)
```
